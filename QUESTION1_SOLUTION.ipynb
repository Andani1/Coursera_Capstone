```python
import urllib.request
```

```python
wiki='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
page=urllib.request.urlopen(wiki)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page)
```

```python
print(soup.prettify)
```

```python
all_tables=soup.find_all('table')
```

```python
right_table=soup.find('table',class_="wikitable sortable")
```

```python
A=[]
B=[]
C=[]


for row in right_table.findAll("tr"):
  
    cells = row.findAll('td')
    if len((cells))==3:
       
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
        
       
        
```

```python
df=pd.DataFrame()
df['PostalCode']=A
df['BOROUGH']=B
df['NEIBOURHOOD']=C

df
```

```python

df.drop(df[df.BOROUGH=='Not assigned'].index,inplace=True)
```

```python
df['NEIBOURHOOD']=np.where(df['NEIBOURHOOD']=='Not assigned',df['BOROUGH'],df['NEIBOURHOOD'])
```

```python
#df2=df.groupby(['PostalCode','BOROUGH'], sort = False).agg(lambda x: ','.join(x))
df2=df.groupby(['PostalCode','BOROUGH'])['NEIBOURHOOD'].apply(', '.join).str.replace('(\\n)', '').reset_index()
df2
```
